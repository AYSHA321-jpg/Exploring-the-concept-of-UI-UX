3. Compare UI vs. UX .

a) List three key differences between UI and UX based on your analysis.

-In PhonePe, UI (User Interface) and UX (User Experience) are distinct but interconnected aspects of the app's design. UI focuses on the visual elements users interact with, like buttons and icons, while UX encompasses the overall user journey and satisfaction. Here are three key differences:

 Focus:
UI design centers on the visual elements and how they appear on the screen, like the layout, color scheme, and button styles. UX design, on the other hand, focuses on the user's overall experience, including how easy it is to navigate, find information, and complete tasks within the app.

 Scope:
UI design is more focused on the specific touchpoints and visual aspects of the app's interface, while UX design considers the entire user journey from the first interaction to the final action. For example, a UI designer might focus on the appearance of a button, while a UX designer would consider how that button fits into the overall flow of a transaction.

 Outputs:
UI designers produce mockups, high-fidelity wireframes, and prototypes that demonstrate the visual design. While UX designers create user personas, journey maps, wireframes, and prototypes that map out the user's interaction with the app.

In essence, UI is about making the app look good, and UX is about making it work well and be enjoyable to use.
